```python
from pylab import *
import pandas as pd
%matplotlib inline
```

```python
tb=pd.read_csv("Data_S1.cdt",sep="\t",header=None,skiprows=7)
```

```python
data=tb.iloc[:,7:].values
imshow(data,aspect="auto",cmap="bwr",vmin=-2.5,vmax=2.5)
colorbar()
xlabel("gene knocked out")
ylabel("transcripts expressed")
```

Condition 
$x>1.7$ or $x<-1.7$ to consider gene to have flipped

```python
hist(sum(abs(data)>1.7,axis=0))
semilogy()
xlabel("number of genes flipped as a result of a knockout")
```
